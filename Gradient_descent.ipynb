```python
import torch

x = torch.tensor([2.,4.,6.], requires_grad=True)
y = torch.tensor([6.,12.,18.], requires_grad=True)

w = torch.tensor(0.0, dtype=torch.float32, requires_grad=True)
iter = 100

learning_rate = 0.01

def forward(x):
    return w * x

def loss_calc(y_pred,y):
    loss = (y_pred - y)**2
    return loss.mean()

for epoch in range(iter):
    
    y_pred = forward(x)
    loss = loss_calc(y_pred, y)
    loss.backward()
    
    with torch.no_grad():
        w -= learning_rate * w.grad
        
    w.grad.zero_()
    
    if epoch % 10 == 0:
        print(f'epoch {epoch+1}: w = {w.item():.3f}, loss = {loss.item():.8f}')
        
print(f'predicted value for f(8) is {forward(8):.3f}')
    
```
